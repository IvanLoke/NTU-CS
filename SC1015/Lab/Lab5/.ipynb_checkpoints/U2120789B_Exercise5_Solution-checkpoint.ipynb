```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn import metrics
from sklearn.metrics import confusion_matrix
```

```python
df = pd.read_csv(r"D:\Work\Y2 work\SC1015\Lab\Lab5\train.csv")
df
```

```python
CentralAir = pd.DataFrame(df['CentralAir'])
SalePrice = pd.DataFrame(df['SalePrice'])
```

```python
CentralAir_train,CentralAir_test,SalePrice_train,SalePrice_test = train_test_split(CentralAir, SalePrice, test_size = 0.2)
```

```python
sb.boxplot(data=df, x="SalePrice",y="CentralAir")
```

## Train dataset for max depth 2

```python
# Import Decision Tree Classifier model from Scikit-Learn
from sklearn.tree import DecisionTreeClassifier

# Create a Decision Tree Classifier object
dectree1 = DecisionTreeClassifier(max_depth = 2)
```

```python
# Train the Decision Tree Classifier model
dectree1.fit(SalePrice_train, CentralAir_train)
```

```python
from sklearn.tree import plot_tree

f = plt.figure(figsize=(12,12))
plot_tree(dectree1, filled=True, rounded=True, 
          feature_names=["SalePrice"], 
          class_names=["N","Y"])
```

## Train dataset for max depth 4

```python
dectree2 = DecisionTreeClassifier(max_depth = 4)
```

```python
dectree2.fit(SalePrice_train, CentralAir_train)
```

```python
from sklearn.tree import plot_tree

salePrice_depth4 = plt.figure(figsize=(36,36))
plot_tree(dectree2, filled=True, rounded=True, 
          feature_names=["SalePrice"], 
          class_names=["N","Y"], fontsize=18)
```

## 1B Confusion Matrix


### Train set for max depth 2

```python
# Predict Legendary corresponding to Total Train
CentralAir_train_pred1 = dectree1.predict(SalePrice_train)

# Print the Classification Accuracy
print("Classification Accuracy \t:", dectree1.score(SalePrice_train, CentralAir_train))
cm = confusion_matrix(CentralAir_train, CentralAir_train_pred1)
tn = cm[0][0]
tp= cm[1][1]
fp = cm[1][0]
fn = cm[0][1]

print(f"The true positive rate is {tp/(tp+fp)}")
print(f"The false positive rate is {fp/(fp+fn)}")
# Plot the two-way Confusion Matrix
from sklearn.metrics import confusion_matrix
sb.heatmap(confusion_matrix(CentralAir_train, CentralAir_train_pred1), 
           annot = True, fmt=".0f", annot_kws={"size": 18})
```

### Train set for max depth 4


```python
# Predict Legendary corresponding to Total Train
CentralAir_train_pred2 = dectree2.predict(SalePrice_train)

# Print the Classification Accuracy
print("Classification Accuracy \t:", dectree2.score(SalePrice_train, CentralAir_train))
cm = confusion_matrix(CentralAir_train, CentralAir_train_pred2)
tn = cm[0][0]
tp= cm[1][1]
fp = cm[1][0]
fn = cm[0][1]

print(f"The true positive rate is {tp/(tp+fp)}")
print(f"The false positive rate is {fp/(fp+fn)}")
# Plot the two-way Confusion Matrix
from sklearn.metrics import confusion_matrix
sb.heatmap(confusion_matrix(CentralAir_train, CentralAir_train_pred2), 
           annot = True, fmt=".0f", annot_kws={"size": 18})
```

### Test set for max depth 2

```python
# Predict Legendary corresponding to Total Train
CentralAir_test_pred1 = dectree1.predict(SalePrice_test)

# Print the Classification Accuracy
print("Classification Accuracy \t:", dectree1.score(SalePrice_test, CentralAir_test))
cm = confusion_matrix(CentralAir_test, CentralAir_test_pred1)
tn = cm[0][0]
tp= cm[1][1]
fp = cm[1][0]
fn = cm[0][1]

print(f"The true positive rate is {tp/(tp+fp)}")
print(f"The false positive rate is {fp/(fp+fn)}")
# Plot the two-way Confusion Matrix
from sklearn.metrics import confusion_matrix
sb.heatmap(confusion_matrix(CentralAir_test, CentralAir_test_pred1), 
           annot = True, fmt=".0f", annot_kws={"size": 18})
```

### Test set for max depth 4

```python
# Predict Legendary corresponding to Total Train
CentralAir_test_pred2 = dectree2.predict(SalePrice_test)

# Print the Classification Accuracy
print("Classification Accuracy \t:", dectree2.score(SalePrice_test, CentralAir_test))
cm = confusion_matrix(CentralAir_test, CentralAir_test_pred2)
tn = cm[0][0]
tp= cm[1][1]
fp = cm[1][0]
fn = cm[0][1]

print(f"The true positive rate is {tp/(tp+fp)}")
print(f"The false positive rate is {fp/(fp+fn)}")
# Plot the two-way Confusion Matrix
from sklearn.metrics import confusion_matrix
sb.heatmap(confusion_matrix(CentralAir_test, CentralAir_test_pred2), 
           annot = True, fmt=".0f", annot_kws={"size": 18})
```

## The classification accuracy for train set depth 4 is better than the classification accuracy for train set depth 2. However, the true positive rate and false positive rate are worse at depth 4 than at depth 2.

## For the test set, depth 2 is better in all aspects of classification accuracy, true positive rate and false positive rate.

## To determine what model is better, it is important to think about which metrics is most important for this dataset. Generally depth 2 will outperform depth 4 if precision matters, ie. if occurance of false positive and false results in general canot be tolerated. If the cost for false positives and false negatives are similar, then we can use classification accuracy. So for this case if air conditioning is a must, then we should look at the false positive rates and true positive rates to ensure that the house they are buying will have airconditioning


## Problem 2

```python
OverallQual = pd.DataFrame(df['OverallQual'])
YearBuilt = pd.DataFrame(df["YearBuilt"])
```

```python
OverallQual_train,OverallQual_test,YearBuilt_train,YearBuilt_test, CentralAir_train,CentralAir_test = train_test_split(OverallQual, YearBuilt, CentralAir, test_size = 0.2)
```

### OverallQual and CentralAir

```python
def decisionTree(xtest_set, ytest_set,name):
    dectree = DecisionTreeClassifier(max_depth = 4)
    dectree.fit(xtest_set, ytest_set)
    f = plt.figure(figsize=(18,18))
    plot_tree(dectree, filled=True, rounded=True, 
              feature_names=[name], 
              class_names=["N","Y"])
    

    
```

```python
def confusionmatrixdraw(xtest_set,ytest_set):
    from sklearn.metrics import confusion_matrix
    dectree = DecisionTreeClassifier(max_depth = 4)
    dectree.fit(xtest_set, ytest_set)
    CentralAir_train_pred = dectree.predict(xtest_set)
    # Print the Classification Accuracy
    print("Classification Accuracy \t:", dectree.score(xtest_set, ytest_set))
    cm = confusion_matrix(ytest_set, CentralAir_train_pred)
    tn = cm[0][0]
    tp= cm[1][1]
    fp = cm[1][0]
    fn = cm[0][1]

    print(f"The true positive rate is {tp/(tp+fp)}")
    print(f"The false positive rate is {fp/(fp+fn)}")
    # Plot the two-way Confusion Matrix
    sb.heatmap(confusion_matrix(ytest_set, CentralAir_train_pred), 
               annot = True, fmt=".0f", annot_kws={"size": 18})
```

## Decision tree for OverallQual train

```python
decisionTree(OverallQual_train,CentralAir_train,"OverallQual")
```

```python
confusionmatrixdraw(OverallQual_train,CentralAir_train)
```

## Confusion matrix for OverallQual train

```python
decisionTree(OverallQual_test,CentralAir_test,"OverallQual")
```

```python
confusionmatrixdraw(OverallQual_test,CentralAir_test)
```

### YearBuilt and CentralAir 

```python
decisionTree(YearBuilt_train,CentralAir_train,"YearBuilt")
```

```python
confusionmatrixdraw(YearBuilt_train,CentralAir_train)
```

```python
decisionTree(YearBuilt_test,CentralAir_test,"YearBuilt")
```

```python
confusionmatrixdraw(YearBuilt_test,CentralAir_test)
```

### The classification accuracy for yearbuilt test set is higher than the classification accuracy for the OverallQual test set. The true positive rate for the yearbuilt test set is lower than the true positive rate for the OverallQual and the false positive rate is higher than the OverallQual.

### The classification accuracy for yearbuilt train is lower than OverallQual train set.The true positive rate is higher and false positive rate is also higher than OverallQual train set.

### Generally, the differences between these two sets are very similar. But since OverallQual did not predict any false positives in the test set, its true positive and false positive rates are not reliable. Hence, I believe that Yearbuilt will be a  more accurate metrics given that its classification accuracy is higher.
